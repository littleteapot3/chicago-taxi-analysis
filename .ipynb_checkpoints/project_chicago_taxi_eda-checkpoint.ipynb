# Analyzing Taxi Ride Patterns in Chicago: Passenger Behavior and Weather Impact


## Introduction

Zuber, a new ride-sharing company preparing to launch in Chicago, has tasked me with analyzing competitor data to understand passenger preferences and how external factors like weather affect ride patterns.

This project is divided into two key objectives:

Exploratory Data Analysis: Investigate which taxi companies and neighborhoods are most active to better understand rider demand.

Hypothesis Testing: Examine whether weather conditions — specifically rain — impact the duration of rides from downtown Chicago (The Loop) to O'Hare International Airport on Saturdays.

Using SQL for data extraction and Python for visualization and statistical testing, I aim to deliver insights that can inform business decisions for Zuber's upcoming launch.

## Key Findings

- **Flash Cab and Taxi Affiliation Services** were the most active companies in mid-November 2017, significantly outperforming others.  
  _→ Zuber can study their coverage areas and operational patterns for competitive insights._


- **High drop-off activity** was concentrated in central neighborhoods like the Loop, Near North Side, and West Loop Gate.  
  _→ These areas may represent strong launch zones due to commuter and tourism demand._


- **No significant difference** was found in ride durations from the Loop to O’Hare between rainy and clear Saturdays.  
  _→ Zuber can likely maintain consistent service expectations for airport routes in varying weather._


- **Data preprocessing is key** when merging ride and weather records.  
  _→ Accurate time alignment and filtering are critical for robust analyses._

# Hypothesis Testing

The hypothesis we're asked to test is: "The average duration of rides from the Loop to O'Hare International Airport changes on rainy Saturdays."

**Forming the Hypothesis:**

- **Null Hypothesis (H0):** The average duration of rides on rainy Saturdays is the same as on non-rainy Saturdays.
- **Alternative Hypothesis (H1):** The average duration of rides on rainy Saturdays is different from that on non-rainy Saturdays.

### Filtering the dataset into rides on rainy (bad weather) Saturdays

```python
# The average duration of rides from the Loop to O'Hare International Airport on rainy Saturdays

# Filter for rainy days
rainy_rides = loop_rides[(loop_rides['weather_conditions'] == 'Bad') & (loop_rides['start_ts'].dt.weekday == 5)]
display(rainy_rides.head())

# The average duration of rides from the Loop to O'Hare International Airport at all other times
sunny_rides = loop_rides[(loop_rides['weather_conditions'] == 'Good') & (loop_rides['start_ts'].dt.weekday == 5)]
display(sunny_rides.head())

# Average duration of Saturday rainy rides in minutes
avg_duration_rainy = rainy_rides['duration_seconds'].mean()/60
print("Average duration of Saturday rainy rides in minutes", avg_duration_rainy)

# Average duration for non-rainy Saturday rides in minutes
avg_duration_sunny = sunny_rides['duration_seconds'].mean()/60
print("Average duration for non-rainy Saturday rides in minutes", avg_duration_sunny)
```

```python
# Hypothesis: "The average duration of rides from the Loop to O'Hare International Airport changes on rainy Saturdays."

# Null Hypothesis (H0): The average duration of rides on rainy Saturdays is the same as on non-rainy Saturdays.
# Alternative Hypothesis (H1): The average duration of rides on rainy Saturdays is different from that on non-rainy Saturdays.

    
# Perform t-test (assuming normal distribution and equal variance)
t_stat, p_value = st.ttest_ind(rainy_rides['duration_seconds'], sunny_rides['duration_seconds'], equal_var=False)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpret results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The average duration of rides on rainy Saturdays is significantly different from non-rainy Saturdays.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the average duration of rides on rainy and non-rainy Saturdays.")

# Add a column to indicate the grouping
rainy_rides['condition'] = 'Rainy'
sunny_rides['condition'] = 'Sunny'


# Combine the DataFrames
df_combined_weather = pd.concat([rainy_rides, sunny_rides])

# Convert duration from seconds to minutes
df_combined_weather['duration_minutes'] = df_combined_weather['duration_seconds'] / 60


# Create a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_combined_weather, x='condition', y='duration_minutes', palette='Set2')
plt.title('Ride Duration on Rainy vs. Clear Saturdays')
plt.xlabel('Weather Condition')
plt.ylabel('Duration (minutes)')
plt.tight_layout()
plt.show()


# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df_combined_weather['duration_minutes'].quantile(0.25)
Q3 = df_combined_weather['duration_minutes'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df_filtered_scatter = df_combined_weather[(df_combined_weather['duration_minutes'] >= lower_bound) & (df_combined_weather['duration_minutes'] <= upper_bound)]

# Create a scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_filtered_scatter, x='start_ts', y='duration_minutes', hue='condition', palette='coolwarm')
plt.title('Ride Duration Over Time (Filtered Saturdays)')
plt.xlabel('Date')
plt.ylabel('Duration (minutes)')
plt.legend(title='Weather Condition')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
